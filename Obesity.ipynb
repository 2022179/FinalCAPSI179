## Exploratory Daxta Analysis(EDA)

```python
df['Age'].hist(bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(
    x='Height',
    y='Weight',
    hue='Gender',
    data=df,
    palette={'Male': 'green', 'Female': 'pink'}
)

plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight by Gender")
plt.legend(title="Gender")
plt.show()
```
